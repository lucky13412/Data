```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('iris.csv')
```

```python
df.columns
```

# PRACTICAL - 1

```python
data = df['sepal_length']
plt.hist(data,bins=25,edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# PRACTICAL -2.A

```python
sns.kdeplot(data=df['sepal_length'],shade=True)
plt.title('Kde plot of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.show()
```

# PRACTICAL - 2.B

```python
sns.kdeplot(data=df,x="sepal_width", y="sepal_length", fill=True, cmap="Blues", shade=True)
plt.title('2D Density Plot of Sepal Width vs Sepal Length\n')
plt.xlabel('Sepal Width')
plt.ylabel('Sepal Length')
plt.show()
```

# PRACTICAL - 3

```python
plt.figure(figsize=(6,3),dpi=100)
plt.title('Count Plot')
sns.countplot(x='species',data=df,palette='Set1');
```

# PRACTICAL - 4

```python
plt.figure(figsize=(6,3),dpi=100)
plt.title('Bar Chart')
sns.barplot(x='species',y='sepal_length',data=df,estimator=np.mean)
plt.ylabel('Mean Sepal length')
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

# PRACTICAL-5

```python
plt.figure(figsize=(12,6),dpi=100)
plt.title('Multiple Bar Chart')
sns.barplot(x='parental level of education',y='math score',data=df,estimator=np.mean,hue='gender',palette='winter_r')
plt.ylabel('Mean Math score')
plt.legend(bbox_to_anchor=(1.15, 1));
```

# PRACTICAL - 6

```python
labels = df['gender'].unique()
values = (df[['gender','math score']].groupby('gender')).mean()['math score'].values
plt.pie(values,labels=labels,autopct='%1.1f%%',startangle=140)
plt.title('Pie Chart');
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
plt.figure(figsize=(6,3))
sns.scatterplot(x='salary',y='sales',data=df,palette='viridis')
plt.title('Scatter Plot salary vs sales\n');
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',palette='viridis')
plt.title('Scatter Plot salary vs sales\n');
```

```python
plt.figure(figsize=(6,3))
sns.boxplot(x='sales',data=df,palette='viridis')
plt.title('Box Plot')
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='level of education',y='sales',data=df,palette='viridis')
plt.title('Box Plot')
```

# PRACTICAL-8

```python
plt.figure(figsize=(6,9),dpi=150)
sns.catplot(x='level of education',y='sales',data=df,kind='violin')
plt.xticks(rotation=90)
plt.title('Violin Plot')
plt.show();
```
